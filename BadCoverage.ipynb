```python
import matplotlib.pyplot as plt
import seaborn as sn

# Create a figure and axes for subplots (4 rows, 2 columns)
fig, axs = plt.subplots(3, 3, figsize=(12, 12))

# Scatter plot for RSRP
sn.scatterplot(x=pre_df['Drive_Lon'], y=pre_df['Drive_Lat'], s=50, alpha=1, palette="Spectral", hue=pre_df['Serving Cell RSRP'], ax=axs[0, 0])
axs[0, 0].set_title('RSRP in each geometry coordinates (dBm)')

# Scatter plot for RSRQ
sn.scatterplot(x=pre_df['Drive_Lon'], y=pre_df['Drive_Lat'], s=50, alpha=1, palette="Spectral", hue=pre_df['Serving Cell RSRQ'], ax=axs[0, 1])
axs[0, 1].set_title('RSRQ in each geometry coordinates (dB)')

# Scatter plot for SINR
sn.scatterplot(x=pre_df['Drive_Lon'], y=pre_df['Drive_Lat'], s=50, alpha=1, palette="Spectral", hue=pre_df['RS SINR Carrier 1'], ax=axs[0, 2])
axs[0, 2].set_title('SINR in each geometry coordinates (dB)')

# Scatter plot for RSSI
sn.scatterplot(x=pre_df['Drive_Lon'], y=pre_df['Drive_Lat'], s=50, alpha=1, palette="Spectral", hue=pre_df['Serving Cell Channel RSSI'], ax=axs[1, 0])
axs[1, 0].set_title('RSSI in each geometry coordinates (dBm)')

# Scatter plot for Throughput DL
sn.scatterplot(x=pre_df['Drive_Lon'], y=pre_df['Drive_Lat'], s=50, alpha=1, palette="Spectral", hue=pre_df['Throughput DL'], ax=axs[1, 1])
axs[1, 1].set_title('Throughput DL in each geometry coordinates (Mbps)')

# Scatter plot for Throughput UL
sn.scatterplot(x=pre_df['Drive_Lon'], y=pre_df['Drive_Lat'], s=50, alpha=1, palette="Spectral", hue=pre_df['Throughput UL'], ax=axs[1, 2])
axs[1, 2].set_title('Throughput UL in each geometry coordinates (Mbps)')

# Scatter plot for CQI
sn.scatterplot(x=pre_df['Drive_Lon'], y=pre_df['Drive_Lat'], s=50, alpha=1, palette="Spectral", hue=pre_df['CQI'], ax=axs[2, 0])
axs[2, 0].set_title('CQI in each geometry coordinates')

# Scatter plot for Distance to Serving Cell
sn.scatterplot(x=pre_df['Drive_Lon'], y=pre_df['Drive_Lat'], s=50, alpha=1, palette="Spectral", hue=pre_df['Dis_Serving'], ax=axs[2, 1])
axs[2, 1].set_title('Distance to Serving Cell (m)')

fig.delaxes(axs[2, 2])

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()



# Save the figure
fig.savefig("BadCoverageCell.png")


```
