```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_res,y_res,test_size=0.2,random_state=42,shuffle=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
s = MinMaxScaler(feature_range=(0,20))
scalded_data_training = s.fit_transform(x_train)
scalded_data_testing = s.transform(x_test)
```

```python
model = models.Sequential()
model.add(Dense(2056,input_dim=7,activation='relu'))
model.add(Dropout(0.5, name = "drop1"))


model.add(Dense(1028,activation='relu'))
model.add(Dropout(0.4, name = "drop2"))

model.add(Dense(512,activation='relu'))
model.add(Dropout(0.2, name = "drop3"))

model.add(Activation('softmax'))
model.compile(loss ='sparse_categorical_crossentropy',optimizer= 'adam',metrics = ['accuracy'])
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
monitors = ModelCheckpoint(filepath='monitor.hdf5',verbose=1,save_best_only=True)
history = model.fit(scalded_data_training,y_train,validation_split=0.2,callbacks=[monitors],verbose=2,epochs=10,shuffle=True)
```

```python
plt.figure()
plt.title('Training performance')
plt.plot(history.epoch, history.history['loss'], label='train loss+error')
plt.plot(history.epoch, history.history['val_loss'], label='val_error')
plt.legend()
```

```python
results = model.evaluate(scalded_data_testing, y_test)
```
