# Cell File Preparation

```python
import pandas as pd
```

```python
CellFile = pd.read_csv("/content/LTE_CellFile.csv")
```

```python
CellFile.shape
```

```python
# TAC Number depends on measurement Area
TAC1 = 42342 # First TAC in Sheikh Zayed Area
TAC2 = 42345 # Second TAC in Sheikh Zayed Area
CellFileAnalysis1 = CellFile['TAC']== TAC1
CellFileAnalysis2 = CellFile['TAC']== TAC2
```

```python
# Filteration of CellFile into Cell File for just the measurement area
CellFile1 = CellFile[CellFileAnalysis1]  
CellFile2 = CellFile[CellFileAnalysis2]
```

```python
# Combine two files together
frames = [CellFile1,CellFile2]
CellFileAnalysisByTac = pd.concat(frames)
```

```python
# To make sure that the shape of file has been minimized
CellFileAnalysisByTac.shape
```

```python
# To correct PCI in range of 0 to 503 as our data measurement file
CellFileAnalysisByTac['CI'] = CellFileAnalysisByTac['PCIG'] * 3 + CellFileAnalysisByTac['PCI'] 
```

```python
# Cell Filter Filteration to CSV File to work with
CellFileAnalysisByTac.to_csv('CellFileFilteration.csv', index=True)
```
